About this file
Dataset Columns:

Person ID: An identifier for each individual.
Gender: The gender of the person (Male/Female).
Age: The age of the person in years.
Occupation: The occupation or profession of the person.
Sleep Duration (hours): The number of hours the person sleeps per day.
Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.
Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.
Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.
BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).
Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.
Heart Rate (bpm): The resting heart rate of the person in beats per minute.
Daily Steps: The number of steps the person takes per day.
Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

# Categorizing columns or creating range of values for the columns

```python
#Getting the unique values of each column excep Person ID and gender and looking for anomalies

unique_data = {}
exclude_columns = ['Person ID', 'Gender']

for column in sleep_data.columns:
    if column not in exclude_columns:
        unique_data[column] = sleep_data[column].unique()

unique_data
```

```python
# We can create range for age,Sleep Duration,Physical activity,Heart Rate ,Daily Steps
# Also Creating a new column based on blood pressure
```

```python
# Split the 'Blood Pressure' column into 'Systolic' and 'Diastolic' columns
sleep_data[['Systolic', 'Diastolic']] = sleep_data['Blood Pressure'].str.split('/', expand=True).astype(int)
```

```python
conditions = [
    (sleep_data['Systolic'] < 120) & (sleep_data['Diastolic'] < 80),
    (sleep_data['Systolic'].between(120, 129)) & (sleep_data['Diastolic'] < 80),
    (sleep_data['Systolic'].between(130, 139)) | (sleep_data['Diastolic'].between(80, 89)),
    (sleep_data['Systolic'] >= 140) | (sleep_data['Diastolic'] >= 90),
    (sleep_data['Systolic'] > 180) | (sleep_data['Diastolic'] > 120)
]

choices = ['NORMAL', 'ELEVATED', 'HIGH BLOOD PRESSURE STAGE 1', 'HIGH BLOOD PRESSURE STAGE 2', 'HYPERTENSIVE CRISIS']
```

```python
sleep_data['Blood Pressure Category'] = np.select(conditions, choices, default='Unknown')

# Display the result
print(sleep_data[['Blood Pressure', 'Blood Pressure Category']])
```
